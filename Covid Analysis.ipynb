### Live COVID-19 Data Tracking: India

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading the dataset
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df=df.drop(index=36,axis=1)
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#state with most active cases
df[df['active']==df['active'].max()]
```

```python
#state with most cured cases
df[df['cured']==df['cured'].max()]
```

```python
#state with most deaths
df[df['death']==df['death'].max()]
```

```python
#state with most new active cases
df[df['new_active']==df['new_active'].max()]
```

```python
#state with most new cured cases
df[df['new_cured']==df['new_cured'].max()]
```

```python
#creading a new column for cured/death ratio
df['cd_ratio']=df['cured']/df['death']
df.head()
```

```python
df=df.replace([np.inf, -np.inf], np.nan)
```

```python
#restructuring the dataset
df=df.drop('sno',axis=1)
df=df.set_index('state_name')
df
```

```python
#performing visualizations
df1=df.iloc[:,1:3]
df1.plot.barh(figsize=(15,15))
plt.show()
```

```python
#plotting state-wise cd ratio
df2=df['cd_ratio'].dropna()
df2.plot.barh(figsize=(15,15),color='orange')
plt.xlabel('Cured/Death')
plt.ylabel('State')
plt.show()
```

```python
#stacked bar chart
df.drop('cd_ratio',axis=1).plot.barh(stacked=True,figsize=(15,15))
plt.show()
```
